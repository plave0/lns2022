# Analiza podataka

## Potrebne biblioteke

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

## Učitavanje i preprocesiranje podataka

```python
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("Podaci/iris.csv")
```

```python
features = data.columns[:-1]
X = data[features]
y = data["Species"]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42, stratify=y)
```

## Klasifikacija

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.tree import plot_tree
```

```python
dt = DecisionTreeClassifier(max_depth=4)
dt.fit(X_train, y_train)
```

```python
train_pred = dt.predict(X_train)
train_pred
```

```python
test_pred = dt.predict(X_test)
test_pred
```

```python
pd.DataFrame(confusion_matrix(y_test, test_pred)
            ,index=dt.classes_
            ,columns=dt.classes_)
```

```python
plt.figure(figsize=(15,15))
plot_tree(dt, feature_names=features, class_names=dt.classes_, filled=False, impurity=False, label=None)
plt.show()
```
